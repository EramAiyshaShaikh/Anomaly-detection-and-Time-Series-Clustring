**Question 1 : What is Dimensionality Reduction? Why is it important in machine learning? **

Answer:
- What is Dimensionality Reduction?
  - It's a technique to simplify complex datasets by decreasing the number of input variables (features).
  - The goal is to create a more concise representation of the data by eliminating redundant or irrelevant features, while keeping the most important information.
  - This is often achieved by transforming the original high-dimensional data into a lower-dimensional space through techniques like feature extraction or feature selection.
- Why is it important in machine learning?
  - Improved efficiency: Reducing the number of features reduces the computational complexity, which leads to faster model training and less data storage requirement.
  - Reduced overfitting: High-dimensional data can lead to overfitting, where a model performs well on training data but poorly on new data. Dimensionality reduction can prevent this by focusing on the most significant patterns.
  - Enhanced visualization: It is difficult to visualize data with many dimensions (e.g., >3). Dimensionality reduction can project the data into 2 or 3 dimensions, making it easier to explore and understand.
  - Better model performance: By removing noisy or irrelevant features, the overall accuracy and performance of the model can be improved.
  - Mitigates the "curse of dimensionality": As the number of features increases, the data becomes sparser, making it harder for algorithms to generalize. Dimensionality reduction helps to counteract this effect.

**Question 2: Name and briefly describe three common dimensionality reduction techniques.**

Answer:
1. Principal Component Analysis (PCA):
PCA is a linear, unsupervised feature extraction technique used for general-purpose dimensionality reduction. It identifies new orthogonal axes (principal components) that maximize the variance in the dataset. The data is then projected onto a smaller subset of these new dimensions that capture most of the original data's variability, effectively reducing dimensionality while retaining essential information.
2. Linear Discriminant Analysis (LDA):
LDA is a linear, supervised dimensionality reduction technique. Unlike PCA's focus on maximizing variance, LDA aims to find a new feature space that maximizes the separation between multiple known classes (maximizing the distance between class means while minimizing the variance within each class). This makes it particularly useful for classification problems.
3. t-Distributed Stochastic Neighbor Embedding (t-SNE):
t-SNE is a non-linear dimensionality reduction technique primarily used for visualizing high-dimensional data in a low-dimensional space (typically 2D or 3D). It focuses on preserving the local structure of the data, meaning that data points which are close together in the original high-dimensional space remain close in the reduced representation. This ability to capture complex, non-linear relationships makes it excellent for revealing clusters in data for visualization purposes.

**Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.**

Answer:
- What is clustering?:
  - Definition: An unsupervised machine learning technique that involves partitioning a dataset into groups, or clusters.Goal: To discover natural groupings in data by organizing objects based on similarity.Process: It aims to create clusters where internal distances are small (similar members) and external distances are large (dissimilar members).

- Three popular clustering algorithms
  - K-means: A centroid-based algorithm that works by partitioning data into k clusters, where each cluster is represented by the mean of the data points within it.
  - Hierarchical clustering: A connectivity-based algorithm that creates a hierarchy of clusters. It can be either agglomerative (bottom-up, merging clusters) or divisive (top-down, splitting clusters).
  - DBSCAN: A density-based algorithm that groups together points that are closely packed together, marking as outliers the points that lie alone in low-density regions.

**Question 4: Explain the concept of anomaly detection and its significance.**

Answer:
- Concept of anomaly detection
  - Identifies unusual data: It involves finding data points that are significantly different from the expected behavior or standard patterns. These anomalies are also known as outliers, novelties, or exceptions.
  - Uses historical data: The process analyzes historical data samples to establish a baseline of normal behavior and then identifies occurrences that don't fit this baseline.
  - Types of anomalies: Anomalies can be categorized based on context:
  1. Point Anomalies: A single data point that is unusual on its own (e.g., a sudden, massive spike in website traffic).
  2. Contextual Anomalies: A data point that is unusual in a specific context (e.g., high internet usage during the middle of the night).
  3. Collective Anomalies: A group of related data points that are collectively unusual, even if individual points seem normal (e.g., multiple failed login attempts in a short period).
- Significance of anomaly detection
  - Cybersecurity: It is crucial for detecting security breaches, network intrusions, and other malicious activities by identifying abnormal patterns in network traffic or user behavior.
  - Fraud prevention: Businesses in sectors like finance use it to detect fraudulent transactions, such as a customer making an unusually large international purchase.
  - Operational stability: It helps monitor infrastructure and applications to identify and address performance anomalies that could lead to system failures or disruptions.
  - Quality assurance: In manufacturing and IoT, it can pinpoint faulty equipment, sensor errors, or production defects that affect product quality.
  - Customer experience: It can be used to identify patterns in customer behavior that signal dissatisfaction or potential churn, allowing companies to intervene proactively.

**Question 5: List and briefly describe three types of anomaly detection techniques.**

Answer:  
1. Statistical Methods:
Statistical methods use mathematical models and tests to flag deviations from expected patterns. They assume data follows a specific distribution (e.g., Gaussian distribution) and identify data points that fall outside a defined range or number of standard deviations from the mean as anomalies.
- Description: These techniques are among the oldest and most trusted, working well with data that fits predictable patterns.
- Example Algorithms: Z-score analysis, Interquartile Range (IQR) analysis, and moving averages.
2. Machine Learning Approaches:
Machine learning (ML) approaches learn patterns of normal behavior from data and flag instances that do not conform to these learned patterns. These methods can be supervised (using labeled normal and anomalous data), unsupervised (using unlabeled data), or semi-supervised (using mostly labeled normal data to build a model of "normal").
- Description: ML methods are highly effective for large, complex, and high-dimensional datasets, as they can adapt to dynamic patterns and uncover subtle anomalies traditional methods might miss.
- Example Algorithms: Clustering algorithms (like K-means or DBSCAN), Isolation Forests, and Autoencoders (a type of neural network).
3. Proximity-Based Methods:
Proximity-based methods operate on the principle that normal data points tend to be close to one another in a data space, while anomalies are located far from their nearest neighbors or in low-density regions.
- Description: These techniques measure the similarity or distance between data points to assess how unusual a specific point is compared to its surroundings.
- Example Algorithms: K-nearest neighbors (k-NN) and Local Outlier Factor (LOF).

**Question 6: What is time series analysis? Mention two key components of time series data.**

Answer:
- Time series analysis is the study of ordered data points collected over time to extract meaningful insights and patterns. Two key components of time series data are trend, which represents the long-term upward or downward movement, and seasonality, which consists of regular, predictable fluctuations that repeat at fixed intervals.
- Key components of time series data
  - Trend: This is the long-term, overall direction of the data. It can show an increasing, decreasing, or stationary pattern over an extended period, such as the upward trend in e-commerce sales over the last five years.
  - Seasonality: This component represents regular, repeating patterns that occur within a specific, fixed timeframe. For example, retail sales consistently spike during the holiday season each year, or electricity usage increases in the summer due to air conditioning.

**Question 7: Describe the difference between seasonality and cyclic behavior in time series.**

Answer:
- Seasonality
  1. Period: Fixed and constant, often tied to a calendar period like daily, weekly, monthly, or quarterly.
  2. Cause: Influenced by factors like weather, holidays, or the time of day (e.g., high traffic during rush hour).
  3. Predictability: Highly predictable because the pattern repeats at exact intervals.
  4. Example: Sales of ice cream peaking every summer, or a daily traffic surge at 5 PM on weekdays.
- Cyclicity
  1. Period: Variable and irregular, typically lasting more than a year.
  2. Cause: Influenced by longer-term economic factors, such as business cycles (expansion and contraction).
  3. Predictability: Less predictable because the length and amplitude of the cycles are not fixed and can vary.
  4. Example: A business experiencing a period of growth followed by a recession over several years.

  
  

```python
'''Question 8: Write Python code to perform K-means clustering on a sample dataset.

Answer:'''

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=300, centers=3, n_features=2, random_state=1)

plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1], s=50, cmap='viridis')
plt.title('Original sample data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans

kmeans.fit(X)

labels = kmeans.labels_

centroids = kmeans.cluster_centers_

plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1], c=labels, s=50, cmap='viridis')
plt.scatter(centroids[:,0],centroids[:,1], marker='x', s=200,color='red', cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

from sklearn.metrics import silhouette_score
sil = silhouette_score(X, labels)
print(f'Silhouette Score: {sil}')
```

```python
'''Question 9: What is inheritance in OOP? Provide a simple example in Python.

Answer:'''
'''Inheritance is an object-oriented programming concept where a new class (child/derived class) inherits properties and methods from an existing class (parent/base class).
This promotes code reusability, as the child class can use the parent's functionality and also add its own unique features.'''
class Vehicle:
  def __init__(self, brand, model, year):
    self.brand = brand
    self.model = model
    self.year = year

  def discription(self):
    print(f"The {self.brand} {self.model} is made in the year {self.year}")

class Car(Vehicle):
  def __init__(self, brand, model, year, doors):
    super().__init__(brand, model, year)
    self.doors = doors

  def display_doors(self):
    print(f"This car has {self.doors} doors.")

my_car = Car("Toyota", "Camry", 2025, 4)
my_car.discription()
my_car.display_doors()
```

**Question 10: How can time series analysis be used for anomaly detection?**

Answer:
- Time series analysis is used for anomaly detection by first modeling the normal behavior and patterns within sequential data and then identifying data points that significantly deviate from this established norm [2]. Key techniques for this include:
1. Modeling Normal Behavior:
The fundamental step involves establishing a baseline of expected values based on historical data. This baseline incorporates patterns such as:
  - Seasonality: Regular, predictable cycles (e.g., increased sales every December) [2].
  - Trends: Long-term upward or downward movements in the data [2].
  - Cyclical Variations: Patterns that repeat over long, sometimes irregular, time periods [2].
Methods like the Moving Average or more advanced statistical models such as ARIMA (AutoRegressive Integrated Moving Average) are often used to forecast future values and define a range of normal outcomes [2, 3].
2. Identifying Anomalies (Outliers):
Once the model is established, anomaly detection involves flagging data points that fall outside the expected range or exhibit unusual behavior. This is typically done through methods such as:
  - Statistical Thresholds: Data points that exceed a certain standard deviation from the predicted mean are marked as anomalies [2].
  - Residual Analysis: Examining the difference (residual) between the actual observed data point and the value predicted by the model [3]. Large residuals indicate potential anomalies.
  - Control Charts: Visual tools, like the ASQ control charts, help monitor the process over time and signal when a point is "out of control" or statistically unlikely to be part of the normal process variation [1].
  - Machine Learning Approaches: Techniques like Isolation Forests or One-Class SVM (Support Vector Machine) are trained on "normal" data to learn boundaries and then identify observations that do not conform to this learned pattern [2].
- In essence, time series analysis distinguishes between expected fluctuations and genuine anomalies by building a predictive model of "normal" and highlighting anything that the model could not reasonably predict [2]
